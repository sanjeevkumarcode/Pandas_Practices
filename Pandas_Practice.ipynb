# 6.a.1. Remove all rows with NULL values 

```python
# It will delete empty cells. It is the 1st way to deal with empty values.

df1.dropna(inplace = True)
df1
```

# 6.a.2. Replace Empty Values

```python
# To handle the missing value, The 2nd way of dealing with empty cells is to insert a new value instead.
# This way you do not have to delete entire rows just because of some empty cells.
# The fillna() method allows us to replace empty cells with a value:

df2=pd.read_excel(r'C:\Users\SANJEEV KUMAR\Desktop\Excel_Practise.xlsx')
df2.fillna(130, inplace = True)
df2
```

```python
m=df3['Population'].mean()
df3['Population'].fillna(m, inplace=True)
df3
```

```python
me=df3['Total Cases'].median()
df3['Total Cases'].fillna(me, inplace=True)
df3
```
